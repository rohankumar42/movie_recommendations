```python
f = open("lmafit_data.out")
```

```python
inp = f.read()
inps = inp.replace(":", ",")
print(inps)
```

```python
inps = inps.split("\n")
```

```python
parsed = [i.split(",") for i in inps]
```

```python
ranks = []
ps = []
rmse = []

for p in parsed:
    if p != ['']:
        ranks.append(float(p[0]))
        ps.append(float(p[1]))
        rmse.append(float(p[2]))
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(10,8))
tog = sorted(zip(ps, ranks, rmse))
togs = np.array([list(t) for t in tog])
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:purple', 'tab:gray']
for i in range(6):
    start_ind = 8 * i
    end_ind = 8 * (i + 1)
    plt.plot(togs[start_ind: end_ind,1],togs[start_ind: end_ind,2] , marker='o', markersize=5, color=colors[i], linewidth=2, label="p=" + str(togs[start_ind][0]))
plt.legend(loc="upper left")
plt.xlabel('Predicted Rank of Matrix')
plt.ylabel('Root Mean Squared Error of LMaFit Completion')
plt.savefig('LMaFit.jpg')
```
